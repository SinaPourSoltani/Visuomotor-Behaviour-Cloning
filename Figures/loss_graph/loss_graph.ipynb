```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
x = list(range(1, 16))

with open("summary_final_baseline_2_0_unfrozen_from_5_17-10-38") as json_file: 
    summary = json.load(json_file)
    train_loss = summary["Train loss"]
    val_loss = summary["Test loss"]
    
    fig = plt.figure(1)
    plt.plot(x, train_loss[:15])
    plt.plot(x, val_loss[:15])    
    plt.plot(x, train_loss[:5] + train_loss[15:25])
    plt.plot(x, val_loss[:5] + val_loss[15:25])



plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.xlim([1, 15])
plt.legend(["Baseline Train", "Baseline Valid", "Unfrozen Train", "Unfrozen Valid"])
fig.savefig('loss_baseline_freeze.png')
```

```python
x = list(range(1, 16))

with open("summary_final_baseline_2_0_unfrozen_from_5_17-10-38") as json_file: 
    summary = json.load(json_file)
    train_loss = summary["Train loss"]
    val_loss = summary["Test loss"]
    
    fig = plt.figure(1)
    plt.plot(x, train_loss[:15], color='C0')
    plt.plot(x, val_loss[:15], color='C1')    
    plt.plot(x, train_loss[:5] + train_loss[15:25], color='C0', linestyle='dashed')
    plt.plot(x, val_loss[:5] + val_loss[15:25], color='C1', linestyle='dashed')



plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.xlim([1, 15])
plt.legend(["Baseline Train", "Baseline Valid", "Unfrozen Train", "Unfrozen Valid"])
fig.savefig('loss_baseline_freeze.png')
```
